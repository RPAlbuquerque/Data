```python
import streamlit as st

# Configura√ß√£o da p√°gina
st.set_page_config(page_title="Mark da Marketri - Quiz de Marketing", page_icon="üìä", layout="centered")

# Cores da marca
COR_PRIMARIA = "#6A0DAD"  # Roxo
COR_SECUNDARIA = "#00BFFF"  # Azul celeste

st.markdown(f"""
    <style>
        .stApp {{
            background-color: #f9f9f9;
        }}
        .title h1 {{
            color: {COR_PRIMARIA};
        }}
        .stButton > button {{
            background-color: {COR_SECUNDARIA};
            color: white;
            font-weight: bold;
        }}
    </style>
""", unsafe_allow_html=True)

# Cabe√ßalho
st.markdown("<h1 style='text-align: center;'>üêµ Mark da Marketri</h1>", unsafe_allow_html=True)
st.markdown("<h3 style='text-align: center;'>Descubra o n√≠vel de Marketing Estrat√©gico da sua empresa</h3>", unsafe_allow_html=True)
st.markdown("---")

pontuacao = 0

# Perguntas
perguntas = [
    {
        "pergunta": "1Ô∏è‚É£ Voc√™ tem um plano de Marketing claro para os pr√≥ximos meses?",
        "alternativas": [
            ("a) N√£o tenho planejamento, fa√ßo postagens e campanhas aleat√≥rias", 1),
            ("b) Tenho algumas ideias, mas sem plano estruturado", 2),
            ("c) Sim, tenho um planejamento bem definido com metas claras", 3)
        ]
    },
    {
        "pergunta": "2Ô∏è‚É£ Voc√™ acompanha e analisa os resultados e dados de sua campanha?",
        "alternativas": [
            ("a) N√£o, s√≥ vejo curtidas e seguidores", 1),
            ("b) Analiso algumas m√©tricas, mas sem muita frequ√™ncia", 2),
            ("c) Uso dados para tomar decis√µes estrat√©gicas e otimizar minhas campanhas", 3)
        ]
    },
    {
        "pergunta": "3Ô∏è‚É£ Seu marketing est√° alinhado com suas metas de vendas?",
        "alternativas": [
            ("a) N√£o, marketing e vendas s√£o coisas separadas pra mim", 1),
            ("b) Tenho algumas a√ß√µes voltadas pra vendas, mas sem funil bem definido", 2),
            ("c) Sim, marketing e vendas andam juntos na minha empresa", 3)
        ]
    },
    {
        "pergunta": "4Ô∏è‚É£ Sua marca tem uma identidade visual forte e reconhec√≠vel?",
        "alternativas": [
            ("a) N√£o tenho um padr√£o visual bem definido", 1),
            ("b) Uso uma identidade visual, mas sem consist√™ncia em todos os canais", 2),
            ("c) Sim, minha identidade visual √© bem estruturada e refor√ßa meu posicionamento", 3)
        ]
    },
    {
        "pergunta": "5Ô∏è‚É£ Voc√™ trabalha com Marketing Virtual e Offline de forma integrada?",
        "alternativas": [
            ("a) Uso s√≥ marketing digital ou offline, n√£o tem integra√ß√£o", 1),
            ("b) Tenho algumas a√ß√µes integradas, mas ainda sem um plano estruturado", 2),
            ("c) Sim, minhas estrat√©gias online e offline s√£o alinhadas para potencializar os resultados", 3)
        ]
    },
]

for i, q in enumerate(perguntas):
    alternativas = [alt[0] for alt in q["alternativas"]]
    escolha = st.radio(q["pergunta"], alternativas, index=None, key=i)
    if escolha:
        pontos = dict(q["alternativas"])[escolha]
        pontuacao += pontos

# Resultado
if st.button("üìä Ver meu n√≠vel de marketing"):
    st.markdown("---")
    if pontuacao <= 10:
        st.error("üîª **N√≠vel: Iniciante**\n\nSeu marketing ainda n√£o tem uma estrat√©gia clara. √â hora de estruturar melhor suas a√ß√µes para colher bons resultados.")
    elif pontuacao <= 20:
        st.warning("üü° **N√≠vel: Intermedi√°rio**\n\nVoc√™ j√° aplica algumas estrat√©gias, mas ainda pode otimizar muito os resultados.")
    else:
        st.success("üîù **N√≠vel: Avan√ßado**\n\nParab√©ns! Seu n√≠vel de marketing estrat√©gico est√° alinhado com o crescimento do seu neg√≥cio.")
```
